```python
#percentage of product type
Coffee = df.loc[df.product_type == 'Coffee Machine']["defect"]
rate_Coffee = sum(Coffee)/len(Coffee)
print(str(rate_Coffee) + " % Product of Coffee Machine." )
```

```python
df['defect'] = df['defect'].map({0:"not_defect",
1: "defect"})
pd.crosstab(index=df['product_type'], columns=df['defect'])
```

```python
df['id_store'] = df['id_store'].map({1:"store_1",
2:"store_2", 3:"store_3", 4:"store_4", 5:"store_5", 6:"store_6", 7:"store_7", 8:"store_8", 9:"store_9",
10:"store_10", 11:"store_11", 12:"store_12"})
pd.crosstab(index=df['quarter'], columns=df['id_store'])
```

```python
sns.histplot(data =df_coffee, x=df_coffee['over_time'], alpha=0.5, color = 'blue')
sns.histplot(data =df_Tea, x=df_Tea['over_time'], alpha=0.5, color = 'red')
```

```python
sns.histplot(data =df_coffee, x=df_coffee['targeted_productivity'], alpha=0.5, color = 'blue')
sns.histplot(data =df_Tea, x=df_Tea['targeted_productivity'], alpha=0.5, color = 'red')
```

```python
ax = sns.histplot(data = df_coffee, x=df_coffee['targeted_productivity'], alpha=0.5, color = 'blue')
ax = sns.histplot(data = df_Tea, x=df_Tea['targeted_productivity'], alpha=0.5, color = 'red')
ax.set_xlabel('targeted_productivity',fontsize = 15)
ax.set_ylabel('Count of records', fontsize = 15)
ax.set_title('Bivariate analysis of targeted_productivity and product_type',
fontsize= 20)
plt.legend(['Coffee_Machine', 'Tea_Machine'],loc="upper right")
```
