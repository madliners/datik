Data Pre-processing

```python
print("\n##  Prapemrosesan Data")
print("--------------------------")
basket = (df.groupby(['ID', 'Items'])['Items']
          .count()
          .unstack(fill_value=0))

def encode_units(x):
    return 1 if x >= 1 else 0

basket_sets = basket.applymap(encode_units)
```

```python
print("5 baris pertama data setelah One-Hot Encoding:")
print(basket_sets.head())
```

Pemodelan Association Mining dengan Algoritma Apriori

```python
min_support_value = 0.3 #spesifikasi min support threshold
min_confidence_value = 0.3  #spesifikasi min confidence threshold
```

```python
print("\n## Algoritma Apriori")
print("--------------------------")
frequent_itemsets = apriori(basket_sets, min_support=min_support_value, use_colnames=True)

print(f"Itemset yang sering muncul (min_support={min_support_value}):")
print(frequent_itemsets.sort_values(by='support', ascending=False).head())
```

```python
print("\n##  Association Rules")
print("--------------------------")
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence_value)

print("Association Rules Lengkap (Diurutkan berdasarkan Lift):")
print(rules.sort_values(by='lift', ascending=False).head())
```

```python
print("\n## Top 3 Rules Berdasarkan Confidence")
print("-----------------------------------------")

# Mengurutkan aturan berdasarkan 'confidence' secara menurun (descending)
rules_sorted_by_confidence = rules.sort_values(by='confidence', ascending=False)

# Mengambil 3 baris pertama
top_3_confidence_rules = rules_sorted_by_confidence.head(3)

print(" 3 Aturan Asosiasi dengan Nilai Confidence Tertinggi:")
print(top_3_confidence_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
### Visualisasi 1: Scatter Plot Support vs. Confidence
plt.figure(figsize=(8, 6))
sns.scatterplot(x="support", y="confidence", data=rules, hue="lift", size="lift", palette="viridis", sizes=(20, 200))
plt.title('Scatter Plot Association Rules: Support vs. Confidence')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.legend(title='Lift', bbox_to_anchor=(1.05, 1), loc=2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show() #
```

```python
### Visualisasi 2: Scatter Plot Support vs. Lift
plt.figure(figsize=(8, 6))
sns.scatterplot(x="support", y="lift", data=rules, hue="confidence", size="confidence", palette="magma", sizes=(20, 200))
plt.title('Scatter Plot Association Rules: Support vs. Lift')
plt.xlabel('Support')
plt.ylabel('Lift')
plt.legend(title='Confidence', bbox_to_anchor=(1.05, 1), loc=2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show() #
```

Force Directed Graph

```python
import networkx as nx
```

```python
# --- Penambahan Program Visualisasi Force-Directed Graph ---
print("\n##  Visualisasi Force-Directed Graph")
print("-----------------------------------------")

# Filter aturan yang memiliki korelasi positif yang kuat (Lift > 1.2)
# Anda dapat menyesuaikan nilai ambang batas ini
rules_for_graph = rules[(rules['lift'] >= 1.2)].copy()

# Konversi frozenset ke string untuk node graph
rules_for_graph['antecedents'] = rules_for_graph['antecedents'].apply(lambda x: ', '.join(list(x)))
rules_for_graph['consequents'] = rules_for_graph['consequents'].apply(lambda x: ', '.join(list(x)))

# Inisialisasi graph
G = nx.DiGraph()

# Tambahkan node (item/itemset) dan edge (aturan)
for index, row in rules_for_graph.iterrows():
    # Menambahkan edge (panah) dari Antecedent ke Consequent
    G.add_edge(row['antecedents'], row['consequents'],
               weight=row['lift'], confidence=row['confidence'])

# Mengatur tata letak (posisi) node
# Menggunakan Spring Layout untuk Force-Directed Graph
pos = nx.spring_layout(G, k=0.5, iterations=50)

# Persiapan untuk pewarnaan dan ukuran node/edge
# Warna Edge berdasarkan Confidence
edges = G.edges()
confidence_values = [G[u][v]['confidence'] for u,v in edges]
edge_colors = plt.cm.plasma(confidence_values)
edge_widths = [G[u][v]['weight'] * 2 for u,v in edges] # Lebar berdasarkan Lift

# Ukuran Node berdasarkan jumlah koneksi (Degree Centrality)
node_sizes = [G.degree(node) * 200 for node in G.nodes()]

plt.figure(figsize=(12, 12))
plt.title(f'Force-Directed Graph Association Rules (Lift >= 1.2)')

# Menggambar Node
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='skyblue', alpha=0.9, linewidths=1, edgecolors='black')

# Menggambar Edge (Panah)
nx.draw_networkx_edges(G, pos, edgelist=edges, width=edge_widths, edge_color=edge_colors,
                       alpha=0.7, arrowstyle='->', arrowsize=20)

# Menggambar Label (Nama Item)
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif', font_weight='bold')

# Membuat Colorbar untuk Confidence
sm = plt.cm.ScalarMappable(cmap=plt.cm.plasma, norm=plt.Normalize(vmin=min(confidence_values), vmax=max(confidence_values)))
sm.set_array([]) # Diperlukan untuk plt.colorbar
cbar = plt.colorbar(sm, ax=plt.gca(), orientation='vertical', fraction=0.03, pad=0.04)
cbar.set_label('Confidence (Warna Edge)')

plt.axis('off') # Sembunyikan sumbu
plt.tight_layout()
plt.show()
```
