```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.plot('day','over_time')
```

```python
df_sub=df.sample(n=100)
```

```python
df_sub.plot('day','over_time')
```

```python
avg =df.groupby('quarter').mean().over_time
batang=avg.plot(kind='bar')
batang.set_ylabel('Average Over Time')
```

```python
plt.scatter(df.no_of_workers,df.actual_productivity,color='C2',facecolor='none')
plt.xlabel('no_of_workers')
plt.ylabel('actual_productivity')
```

```python
histogram=df.actual_productivity.hist()
histogram.set_xlabel('actual_productivity')
histogram.set_ylabel('freq')
```

```python
df.boxplot(column='over_time',by='product_type')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)
df.boxplot(column='over_time', by='product_type', ax=axes[0])
df.boxplot(column='incentive', by='product_type', ax=axes[1])
df.boxplot(column='actual_productivity', by='product_type', ax=axes[2])
for ax in axes :
  ax.set_xlabel('product_type')
  plt.suptitle('')
  plt.tight_layout()
```

```python
korelasi=df.corr()
sns.heatmap(korelasi,xticklabels=korelasi.columns,yticklabels=korelasi.columns,vmin=-1,cmap="RdBu")
fig, ax = plt.subplots()
fig.set_size_inches(11, 7)
sns.heatmap(korelasi, annot=True, fmt=".1f", cmap="RdBu", center=0, ax=ax)
```
